```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['sin(x)'])
plt.grid()
# plt.show()
```

```python
# fig = plt.figure()
# ax = fig.add_subplot(111)
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(['sin(x)'])
ax.grid()
# plt.show()
```

```python
fig, ax = plt.subplots(2, 1)

x = np.arange(0, 4*np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
z = y1 + y2

ax[0].plot(x, y1, ls='-', label='sin', color='b')
ax[0].plot(x, y2, ls='-.', label='cos', color='r')
ax[0].legend()
ax[0].grid()
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_xlim(0, 4*np.pi)

ax[1].plot(x, z, label='sin+cos', color='purple', marker='.')
ax[1].legend(loc='lower left')
ax[1].grid(ls=':')
ax[1].set_xlabel('x')
ax[1].set_ylabel('sin+cos')
ax[1].set_xlim(0, 4*np.pi)

fig.tight_layout()
```

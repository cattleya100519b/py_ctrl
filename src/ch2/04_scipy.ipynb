```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

# 
def system(y, t):
    if t < 10.0:
        u = 0.0
    else:
        u = 1.0
    dydt = (-y + u)/5.0
    return dydt

# 
y0 = 0.5

# 
t = np.linspace(0, 40, 1000)

# 
y = odeint(system, y0, t)

# 
fig, ax = plt.subplots()
ax.plot(t, y, label='y(t)')
# ax.plot(t, [1.0 if i >= 10.0 else 0.0 for i in t], label='u(t)')
ax.plot(t, 1*(t>=10), label='u(t) (step function)', ls='--')
ax.legend(loc='best')
ax.set_xlabel('Time')
ax.set_ylabel('y, u')
ax.grid(ls=':')

print('https://chatgpt.com/share/67ada7c7-2034-8008-8caa-613d5d302c84')

from common import disp_tex

disp_tex(r'\begin{cases} \frac{dy}{dt} = -y(t) + \frac{u}{5} \\ y(0) = 0.5 \end{cases}')
disp_tex(r'\begin{cases} u = 0 & (t < 10) \\ u = 1 & (t \ge 10) \end{cases}')
disp_tex(r'y(t) = \begin{cases} y(0)e^{-t/5} & (t < 10) \\ 1+(y0-e^2)e^{-t/5} & (t \ge 10) \end{cases}')

print(y.min())
print(y0 * np.exp(-(1/5)*10))
print(1+(y0 - np.exp(2)) * np.exp(-(1/5)*10))
print(y.max())
print(1+(y0 - np.exp(2)) * np.exp(-(1/5)*40))
```
